```python
import numpy as np
import matplotlib.pyplot as plt

m = np.linspace(1,2,2)
n = np.linspace(1,2,2)
x = np.linspace(0, 2 * np.pi, 1000)
def func(m,n):
    return np.sinh(np.sqrt((np.pi * m)**2 + (np.pi * n)**2) * x)

labels = [['',''],['','']]
results = np.zeros((len(m), len(n), len(x)))
for i in range(len(m)):
	for j in range(len(n)):
		results[int(i),int(j)] = func(i, j)
		labels[i][j] = f'm = {m[i]}, n = {n[j]}'

labels = np.concatenate(labels).tolist()
results = np.concatenate(results)

for _ in range(1,5):
	plt.subplot(2, 2, _)
	plt.plot(x, results[_-1], label= labels[_-1])
	plt.legend()


plt.show()
```

```python
import numpy as np
import scipy.constants as cons

c = cons.speed_of_light
pi = np.array([10e3, 140]) # in MeV
m_W = np.array([300e3, 80.4e3]) # in MeV
m_t = np.array([300e3, 172.5e3]) # in MeV

def par_mom(E, mass):
    return 1/E - 1/2 * (np.pow(mass,2) / np.pow(E,2))


def rapidity(E, m):
    return 1/2 * np.log( (E + par_mom(E,m) )/ (E - par_mom(E,m)))

rapidities = [rapidity(pi[0], pi[1]), rapidity(m_W[0], m_W[1]), rapidity(m_t[0], m_t[1])]

print(rapidities)
```
