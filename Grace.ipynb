### A .1. b)

Gedankengang: 

- Als erstes die zweit Punkte $P = (5 | f(5))$ und $Q = (10|f(10))$ bestimmen
- Länge einfach durch $|\vec Q - \vec P|$ gegeben
- Um die vertikale Höhe zu bestimmen, wird eine Funktion durch $\vec P$ und $\vec Q$ gesucht. $f(x)$ wird dann von $g(x)$ abgezogen um die Höhe zu erhalten. 

Gleichung $g(x)$:

$$
g(x) = x \cdot \frac{f(10) - f(5)}{10 -5} + c
$$


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
	return 0.0008 * x**4 - 0.12 * x**2 +5

x = np.linspace(0,11,1000)
P = (5,f(5))
Q = (10,f(10))

plt.plot(x, f(x), label='f(x)')
plt.plot(P[0], P[1], 'ro')
plt.plot(Q[0],Q[1], 'ro')

def g(x):
	return x * (Q[1] - P[1])/(Q[0] - P[0]) + 4

plt.plot(x,g(x), label= 'g(x)')

plt.grid()
plt.legend()
plt.show()
```
